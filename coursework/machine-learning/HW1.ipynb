```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. 載入資料集
penguins = sns.load_dataset('penguins')

# 2. 確認資料集的基本資訊（可以檢查是否有遺漏值等）
print(penguins.info())
print(penguins.head())

# 3. 資料清理：如果有遺漏值（NaN），先將其去除
penguins.dropna(inplace=True)

# 4. 設定特徵 X 與目標 y
#   假設我們想要預測「species」（企鵝的種類），
#   那麼其餘欄位就可以視為特徵。
X = penguins.drop('species', axis=1)
y = penguins['species']

# 5. 為了讓機器學習模型能正確處理字串型態的特徵（如 island, sex），
#   可以進行必要的編碼處理。例如利用 pandas 的 get_dummies:
X = pd.get_dummies(X, columns=['island', 'sex'])

# 6. 分割訓練集與測試集
#   依照作業要求，test_size=0.2, random_state=1
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.2, 
    random_state=1
)

# 檢查分割後的資料大小
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```
